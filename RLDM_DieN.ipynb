```python
#################
# DO NOT REMOVE
# Versions
# numpy==1.18.0
################
import numpy as np

class MDPAgent(object):
    def __init__(self):
        pass
    
    #calculate expected loss
    def expected_loss(self,dollars,is_bad_side, N_side):
        return sum(is_bad_side)*1.0/N_side*dollars

    #calculate expected gain
    def expected_gain(self,is_bad_side):
        tot_gain=0
        for idx,isbad in enumerate(is_bad_side):
            tot_gain += (idx+1)*(1-isbad)
        return tot_gain*1.0/len(is_bad_side)

    #check whether the game will continue 
    def continue_game(self, dollars, is_bad_side, N_side):
        gain=self.expected_gain(is_bad_side)
        loss=self.expected_loss(dollars,is_bad_side, N_side)
        return gain>loss

    

    #calculate expected return
    def solve(self, is_bad_side):
        """Implement the agent"""        
        N_side=len(is_bad_side)
        state_value=0 #initial return
        count=0  #initial rounds
        valid_states=[0] #current states(money amount) that the game will continue
        next_valid_states=[] #next states that the game will continue
        exp_gain=self.expected_gain(is_bad_side)
        pre_state_value = -1
        lossProba = sum(is_bad_side)*1.0/len(is_bad_side)
        while True:
            chance=1/N_side**count
            chance_exp_gain = chance * exp_gain 
            for state in valid_states:
                exp_loss=self.expected_loss(state,is_bad_side, N_side)
                state_value += chance_exp_gain - chance*exp_loss
                #iterate through all elements in is_bad_side to find states that allow the game 
                #to continue in next while loop
                for idx,isbad in enumerate(is_bad_side):
                    dollars=state+idx+1
                    if isbad==0 and lossProba*(dollars) < exp_gain:
                        next_valid_states.append(dollars)
            #if the game cannot continue
            if not next_valid_states:
                return state_value
            #if the game continues, get valid_states for next while loop from next_valid_states 
            else:
                valid_states=next_valid_states
                next_valid_states=[]
                count +=1
                if abs(state_value- pre_state_value)<0.001:
                    return state_value
                pre_state_value = state_value
        return state_value
        
        #pass
        #return True
```
